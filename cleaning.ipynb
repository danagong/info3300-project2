```python
trees = pd.read_csv("data/2015_Street_Tree_Census_-_Tree_Data.csv")
trees.head()
```

```python
trees_alive = trees.loc[trees['status'] == 'Alive']
trees_alive = trees_alive.loc[trees['health'] == 'Good']
trees_alive = trees_alive.loc[trees['tree_dbh'] > 20]
trees_alive.head()
```

```python
air = pd.read_csv("Air_Quality.csv")
air.head()
```

```python
air.tail()
```

```python
air['Name'].unique()
```

```python
particulates = air.loc[air['Name'] == 'Fine Particulate Matter (PM2.5)']
particulates.head()
```

```python
particulates.shape
```

```python
particulates['Geo Place Name']
```

```python
particulates['Geo Type Name'].unique()
particulates = particulates.loc[particulates['Geo Type Name'] == 'CD']
particulates.head()
```

```python
particulates.shape
```

```python
particulates['Geo Join ID']
```

```python
particulates = particulates.groupby('Geo Join ID')['Data Value'].mean().reset_index()
particulates = particulates.rename(columns={'Geo Join ID': 'community_district', 'Data Value': 'particle_concentration'})
```

```python
particulates.to_csv('air.csv') 
```
